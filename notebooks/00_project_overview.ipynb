# Storm Frequency–Driven Flood Risk Framework (Overview)

This notebook provides a guided overview of a replicable workflow linking:

1. **Observed extreme rainfall frequency** (PRISM)
2. **Epoch-based change detection** (historical comparison windows)
3. **Projected frequency change** (CORDEX)
4. **Flood probability context** (interpretive, non-deterministic)
5. **Population exposure and vulnerability** (census-based metrics)

The goal is to translate climate non-stationarity into decision-relevant risk insights for emergency management, planning, and environmental justice applications.

## What you need

This repository does not redistribute restricted or large datasets. To run the full workflow on real data, you will need:

- PRISM daily precipitation data (or a processed daily CSV)
- CORDEX daily precipitation output (NetCDF)
- Census/ACS tables and geographic boundaries (or exported GIS join outputs)
- Flood hazard context data (e.g., FEMA flood zones, gauge metadata)

Small demo datasets are included so the notebooks can run out-of-the-box.

## What you get

Running the workflow produces:

- Annual extreme rainfall event counts (PRISM)
- Historical frequency change metrics (epoch comparisons)
- Directional signals of future frequency change (CORDEX)
- Tables and maps showing where exposure and vulnerability concentrate

```python
# ----------------------------------
# Repo-aware path setup (REQUIRED)
# Matches notebook_03
# ----------------------------------
import sys
from pathlib import Path
import pandas as pd

# Resolve repository root (notebooks/ is one level down)
REPO_ROOT = Path("..").resolve()

# Ensure src/ is importable (future-proofing)
if str(REPO_ROOT) not in sys.path:
    sys.path.insert(0, str(REPO_ROOT))

# Core repo directories
DATA_DIR = REPO_ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
DATA_SAMPLE_DIR = DATA_DIR / "sample"

OUTPUTS_DIR = REPO_ROOT / "outputs"
TABLES_DIR = OUTPUTS_DIR / "tables"
MAPS_DIR = OUTPUTS_DIR / "maps"

DOCS_DIR = REPO_ROOT / "docs"
SAMPLE_DIR = REPO_ROOT / "sample_data"

# Ensure output directories exist
TABLES_DIR.mkdir(parents=True, exist_ok=True)
MAPS_DIR.mkdir(parents=True, exist_ok=True)

print("Repo root:", REPO_ROOT)
print("Data dir:", DATA_DIR)
print("Tables dir:", TABLES_DIR)
print("Maps dir:", MAPS_DIR)
```

## How this notebook works

This overview notebook is designed to **read outputs generated by the module notebooks**, rather than rerunning every computation.

Notebook sequence:

## 1) Observed Extreme Rainfall Frequency (PRISM)

Extreme rainfall events are defined using a fixed daily threshold (e.g., ≥ 3.0 inches in 24 hours).
Annual event counts are used to assess how frequency has changed over time.

```python
prism_counts_path = TABLES_DIR / "prism_annual_extreme_counts.csv"

if prism_counts_path.exists():
    prism_counts = pd.read_csv(prism_counts_path)
    display(prism_counts.head())
else:
    print("Missing:", prism_counts_path)
    print("Run notebooks/02_prism_extremes_analysis.ipynb to generate this table.")
```

## 2) Epoch Comparison (Historical Change)

To avoid assuming stationarity, event frequency is compared across multi-year historical windows (epochs), such as:

- 1980–2000
- 2001–2021

```python
epoch_path = TABLES_DIR / "epoch_comparison_summary.csv"

if epoch_path.exists():
    epoch = pd.read_csv(epoch_path)
    display(epoch.head())
else:
    print("Missing:", epoch_path)
    print("Run notebooks/03_epoch_comparison.ipynb to generate this table.")
```

## 3) Projected Frequency Change (CORDEX)

The same frequency-based framing is applied to CORDEX daily precipitation output to evaluate directional changes in extreme-event frequency.

Notebook 04 saves **gridded outputs** (NetCDF) and an optional map image (PNG).

```python
cordex_early = TABLES_DIR / "cordex_future_extreme_freq_early.nc"
cordex_late  = TABLES_DIR / "cordex_future_extreme_freq_late.nc"
cordex_diff  = TABLES_DIR / "cordex_future_extreme_freq_diff.nc"
cordex_png   = MAPS_DIR / "cordex_future_extreme_freq_diff.png"

print("CORDEX outputs present?")
print(" - early:", cordex_early.exists(), "|", cordex_early)
print(" - late :", cordex_late.exists(), "|", cordex_late)
print(" - diff :", cordex_diff.exists(), "|", cordex_diff)
print(" - map  :", cordex_png.exists(), "|", cordex_png)

if not (cordex_early.exists() and cordex_late.exists() and cordex_diff.exists()):
    print("\nRun notebooks/04_cordex_future_frequency.ipynb to generate these outputs.")
```
