```python
# Storm Frequency–Driven Flood Risk Framework (Overview)

This notebook is a guided walkthrough of a replicable workflow linking:

1) **Observed extreme rainfall frequency** (PRISM)  
2) **Epoch-based change detection** (historical comparison windows)  
3) **Projected frequency change** (CORDEX)  
4) **Flood hazard context** (river / floodplain interpretation)  
5) **Population exposure & vulnerability** (census-based metrics)

The goal is to translate climate non-stationarity into decision-relevant risk insights for emergency management, planning, and environmental justice applications.
```

```python
## What you need
This repo does not redistribute restricted or large datasets. You will need:
- PRISM daily precipitation (or processed county time series)
- CORDEX daily precipitation NetCDF (or processed exceedance counts)
- Census/ACS tables and boundaries (or exported join outputs)
- Flood hazard context data (e.g., FEMA polygons, gauge metadata)

## What you get
- Extreme rainfall event frequency and change metrics
- Maps/tables showing where frequency increases concentrate
- Exposure metrics linking hazard context to vulnerability indicators
```

```python
from pathlib import Path
import pandas as pd

# Repo-aware paths
REPO_ROOT = Path("..").resolve()      # notebooks/ is one level down
DATA_DIR = REPO_ROOT / "data"
OUTPUTS_DIR = REPO_ROOT / "outputs"
DOCS_DIR = REPO_ROOT / "docs"

# Expected subfolders
SAMPLE_DIR = REPO_ROOT / "sample_data"   # if you use sample_data/
DATA_SAMPLE_DIR = DATA_DIR / "sample"    # if you use data/sample/

print("Repo root:", REPO_ROOT)
print("Outputs:", OUTPUTS_DIR)
```

```python
## How this notebook works
This overview notebook is designed to **read outputs generated by the module notebooks**, rather than rerunning every computation.

Modules:
- `01_prism_extremes.ipynb` → generates annual event counts
- `02_epoch_comparison.ipynb` → generates epoch comparison tables
- `03_future_frequency.ipynb` → generates CORDEX difference grids or summaries
- `04_flood_context_gauges.ipynb` → generates flood context summaries
- `05_vulnerability_overlay.ipynb` → generates exposure/vulnerability tables


If you havent run those yet, this notebook can still be read as narrative documentation.
```

```python
# 1) Observed Extreme Rainfall Frequency (PRISM)

We define an extreme rainfall event using a fixed threshold (example: **≥ 3 inches in 24 hours**).  
We then count events annually and summarize how frequency changes across historical windows.
```

```python
# Expected output file from PRISM notebooks (you'll generate this later)
prism_counts_path = OUTPUTS_DIR / "tables" / "prism_annual_extreme_counts.csv"

if prism_counts_path.exists():
    prism_counts = pd.read_csv(prism_counts_path)
    display(prism_counts.head())
else:
    print("Missing:", prism_counts_path)
    print("Run notebooks/01_prism_extremes.ipynb to generate this table.")
```

```python
# 2) Epoch Comparison (Historical Change)

To avoid assuming stationarity, we compare event frequency across time windows (“epochs”), such as:
- 1980–2000 vs. 2001–2021

This produces percent change metrics that are interpretable for planning and risk communication.
```

```python
epoch_path = OUTPUTS_DIR / "tables" / "epoch_comparison_summary.csv"

if epoch_path.exists():
    epoch = pd.read_csv(epoch_path)
    display(epoch.head())
else:
    print("Missing:", epoch_path)
    print("Run notebooks/02_epoch_comparison.ipynb to generate this table.")
```

```python
# 3) Projected Frequency Change (CORDEX)

We apply the same frequency framing to CORDEX daily precipitation output:
- Convert to daily exceedance (event / no event)
- Sum exceedances across two future periods
- Compute the difference (late future minus early future)

This yields directional signals of where extreme-event frequency may increase.
```

```python
cordex_summary_path = OUTPUTS_DIR / "tables" / "cordex_future_frequency_summary.csv"
cordex_map_path = OUTPUTS_DIR / "maps" / "cordex_frequency_change.png"

if cordex_summary_path.exists():
    cordex = pd.read_csv(cordex_summary_path)
    display(cordex.head())
else:
    print("Missing:", cordex_summary_path)
    print("Run notebooks/03_future_frequency.ipynb to generate this table (or map).")

if cordex_map_path.exists():
    from PIL import Image
    display(Image.open(cordex_map_path))
else:
    print("Missing:", cordex_map_path)
```

```python
# 4) Flood Context (River / Floodplain Interpretation)

Storm frequency increases matter because they can:
- compress return periods (so “rare” floods occur more often),
- reduce recovery time between events,
- increase cumulative exposure.

This module provides contextual interpretation using gauge-based summaries and flood hazard indicators.
```

```python
flood_context_path = OUTPUTS_DIR / "tables" / "flood_context_summary.csv"

if flood_context_path.exists():
    flood_context = pd.read_csv(flood_context_path)
    display(flood_context.head())
else:
    print("Missing:", flood_context_path)
    print("Run notebooks/04_flood_context_gauges.ipynb to generate this table.")
```

```python
# 5) Exposure & Vulnerability Overlay (Census)

We link hazard context to population vulnerability indicators (race/ethnicity, age, disability, socioeconomic status) to understand:
- who is most likely to be repeatedly exposed,
- where cumulative risk concentrates.

GIS join logic is documented in: `docs/gis_operations_log.md`.
```

```python
vuln_path = OUTPUTS_DIR / "tables" / "vulnerability_exposure_summary.csv"

if vuln_path.exists():
    vuln = pd.read_csv(vuln_path)
    display(vuln.head())
else:
    print("Missing:", vuln_path)
    print("Run notebooks/05_vulnerability_overlay.ipynb to generate this table.")
```
