```python
# PRISM Precipitation Ingestion and Processing

This notebook prepares observed daily precipitation data from PRISM for extreme rainfall frequency analysis under climate non-stationarity.

Extreme rainfall events are defined consistently across time as **daily precipitation ≥ 3.0 inches; 77 mm**.  
(Threshold value can be adjusted by the user; see configuration below.)

---

## Purpose

- Download or ingest daily PRISM precipitation data
- Subset data to a defined region (e.g., counties of interest)
- Clean and standardize time series for downstream extreme-event analysis

This notebook focuses on **data preparation only**.  
Event counting and frequency analysis occur in subsequent notebooks.

---

## Inputs

- PRISM daily precipitation data (user-provided)
- County or regional boundary geometry (for spatial subsetting)

---

## Outputs

- Cleaned daily precipitation time series
- County- or region-level precipitation tables suitable for:
  - extreme event flagging
  - epoch comparison
  - replication in other environments

---

## Notes

- This notebook does **not** redistribute PRISM data.
- Users are responsible for complying with PRISM data use policies.
- All file paths are local and relative to the repository structure.
```

```python
from pathlib import Path
import pandas as pd

print("Notebook running successfully.")
```

```python
# -----------------------------
# USER CONFIGURATION
# -----------------------------

# Path to PRISM CSV or extracted data
PRISM_DATA_PATH = Path("../data/raw/prism_daily_precip.csv")

# Name of date and precipitation columns
DATE_COL = "Date"
PRECIP_COL = "inchesPpt"

# Extreme rainfall threshold (inches)
EXTREME_THRESHOLD_IN = 3.0

# Optional: restrict to specific counties
COUNTY_FILTER = None  # e.g., ["Buncombe", "Haywood"]
```

```python
if not PRISM_DATA_PATH.exists():
    raise FileNotFoundError(
        f"PRISM data not found at {PRISM_DATA_PATH}. "
        "Please download PRISM data and place it in data/raw/."
    )
```

```python
from src.prism_extremes import load_prism_timeseries

df = load_prism_timeseries(
    PRISM_DATA_PATH,
    date_col=DATE_COL,
    precip_col=PRECIP_COL
)

df.head()
```

```python
print("Date range:", df.index.min(), "to", df.index.max())
print("Missing values:", df.isna().sum())
```

```python
OUTPUT_PATH = Path("../outputs/tables/prism_daily_cleaned.csv")
OUTPUT_PATH.parent.mkdir(parents=True, exist_ok=True)

df.to_csv(OUTPUT_PATH)

print(f"Cleaned PRISM data saved to {OUTPUT_PATH}")
```
